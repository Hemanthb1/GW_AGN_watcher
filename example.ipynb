```python
#obtain all interesting candidates through skymap links from urls3 array (O4c run)
import pandas as pd
final = []

for x in urls3:
    try:
        cand, ra, dec, url = run_pipeline(x, file)

        if cand is None or cand.empty:
            continue

        cand["event_url"] = url
        final.append(cand)

    except Exception as e:
        print(f"Failed for {x}: {e}")
        continue

final = pd.concat(final, ignore_index=True) if final else pd.DataFrame()
print(final)
```
